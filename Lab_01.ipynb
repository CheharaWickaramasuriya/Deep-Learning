```python
import numpy as np
exp_arr = np.random.exponential(scale=1.0, size=(4, 4))
print(exp_arr)
```

```python
import matplotlib.pyplot as plt

plt.hist(np.random.exponential(0.1,100000), density=True, bins=50, histtype="step", color="green", label="exponential")
plt.hist(np.random.rand(100000), density=True, bins=100, histtype="step", color="blue", label="uniform")
plt.hist(np.random.randn(100000), density=True, bins=100, histtype="step", color="red", label="normal")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, linewidth=0.1)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

stats_df = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]

# ✅ Pearson Correlation
pearson_corr = stats_df.corr(method='pearson')
plt.figure(figsize=(8, 6))
sns.heatmap(pearson_corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Pearson Correlation Heatmap")
plt.show()

# ✅ Spearman Correlation
spearman_corr = stats_df.corr(method='spearman')
plt.figure(figsize=(8, 6))
sns.heatmap(spearman_corr, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title("Spearman Correlation Heatmap")
plt.show()
```

```python
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
new_df.head(20)
```

```python
# Mean
print("Mean values:")
print(new_df.mean())

# Standard deviation
print("\nStandard deviation:")
print(new_df.std())

# Quartiles
print("\nQuartile values:")
print(new_df.describe().loc[['25%', '50%', '75%']])
```

```python
plt.figure(figsize=(20, 4))

for i, col in enumerate(new_df.columns):
    plt.subplot(1, 4, i+1)
    sns.histplot(new_df[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(new_df)
plt.show()
```

```python
g = sns.PairGrid(new_df)
# Upper triangle
g.map_upper(sns.histplot)

# Diagonal
g.map_diag(sns.histplot, kde=True)

# Lower triangle
g.map_lower(sns.kdeplot, fill=True)

plt.show()
```
